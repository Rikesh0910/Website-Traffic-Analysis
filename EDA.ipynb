# Exploratory Data Analysis

```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
file_path = "C:\\Users\\ADMIN\\Desktop\\Website Traffic Analysis\\Data\\website_data.csv"

df = pd.read_csv(file_path)
```

```python
df.rename(columns={"Session Id": "session_id"}, inplace=True)
df.rename(columns={"Page Views": "page_views"}, inplace=True)
df.rename(columns={"Session Duration": "session_duration"}, inplace=True)
df.rename(columns={"Bounce Rate": "bounce_rate"}, inplace=True)
df.rename(columns={"Traffic Source": "traffic_source"}, inplace=True)
df.rename(columns={"Time on Page": "time_on_page"}, inplace=True)
df.rename(columns={"Previous Visits": "previous_visits"}, inplace=True)
df.rename(columns={"Conversion Rate": "conversion_rate"}, inplace=True)
```

```python
df.head()
```

```python
df["pages_per_min"] = df['page_views'] / (df['session_duration'] / 60)
```

```python
df.head()
```

```python
df.drop(columns=["pages_per_min"], inplace=True)
```

```python
df.head()
```

```python
df["user_type"] = df["previous_visits"].apply(lambda x: "New Visitor" if x == 0 else "Returning Visitor")
```

```python
df.head()
```

```python
df["conversion_status"] = np.where(df["conversion_rate"] == 1, "Converted", "Not Converted")
```

```python
df.head()
```

```python
df["conversion_status"].value_counts()
```

```python
df["time_on_page"].describe()
```

```python
df["bounce_rate_pct"] = df["bounce_rate"] * 100
```

```python
df.head()
```

```python
df["traffic_source"].value_counts()
```

```python
df.groupby("traffic_source")["user_type"].value_counts()
```
